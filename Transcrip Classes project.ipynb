# Transcribe Classes with whisper

```python
import whisper
model = whisper.load_model("base.en")
```

```python
result = model.transcribe(
    r"C:\Users\Alexis\AppData\Local\CapCut\Videos\0411.mp3",
    verbose=True,
    word_timestamps=True,
    no_speech_threshold=0.0,              # Force decode even if silence
    condition_on_previous_text=False      # Prevent early stopping or repetition
)
```

```python
import csv
from datetime import timedelta

# Helper to format seconds as MM:SS.mmm
def format_timestamp(seconds: float) -> str:
    td = timedelta(seconds=seconds)
    total_minutes, secs = divmod(td.total_seconds(), 60)
    return f"{int(total_minutes):02d}:{secs:06.3f}"

# Write segment-level CSV
with open("transcription_segments.csv", "w", encoding="utf-8", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["start", "end", "text"])  # Header

    for segment in result["segments"]:
        start = format_timestamp(segment["start"])
        end = format_timestamp(segment["end"])
        text = segment["text"].strip()
        writer.writerow([start, end, text])
```
